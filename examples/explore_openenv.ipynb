````python
import sys
from pathlib import Path

# Add the parent of argdown_analysis to path (so Python can find "argdown_analysis" folder)
sys.path.insert(0, str(Path("../src/envs").resolve()))

from argdown_analysis import ArgdownAnalysisAction, ArgdownAnalysisEnv
````

````python

# Automatically start container and connect
client = ArgdownAnalysisEnv.from_docker_image("ghcr.io/debatelab/openenv-argdown-analysis:latest")

# Reset the environment
result = client.reset()
print(result.observation.prompt)  
````

````python
analysis = """\
```argdown
[Social media]: Social media is bad.
  + [Loneliness]: Social media increases loneliness.
    + [Face-to-face interaction]: Social media reduces face-to-face interaction.
  - [Cosmopoloitanism]: Social media increases cosmopolitanism.
```
"""

# Send messages
result = client.step(ArgdownAnalysisAction(message=analysis))
print(result.reward) 
print(result.observation.prompt)  
````

````python
client.close()
````
